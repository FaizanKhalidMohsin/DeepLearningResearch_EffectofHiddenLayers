```python
data = pd.read_csv('ted_main.csv')
data2 = data[["event","speaker_occupation","tags"]]
```

```python
tags = data2.tags.str.cat(sep=', ')
tags = tags.replace("'","")
wordcloud = WordCloud(background_color="white").generate(tags)
plt.figure(figsize=[7,7])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
